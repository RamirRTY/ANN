```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_gen = ImageDataGenerator(rescale=1/255)
test_gen = ImageDataGenerator(rescale=1/255)
```

```python
 train_ganator = train_gen.flow_from_directory(directory=basepath , batch_size=128 , class_mode= "binary" , target_size=(300 , 300)  )
 test_ganator = test_gen.flow_from_directory(directory=basepath2 , batch_size=128 , class_mode= "binary" , target_size=(300 , 300)  )
```

```python
 train_ganator.class_indices
```

```python
from keras.applications.vgg16 import VGG16
```

```python
 t_model = VGG16(include_top=False , weights="imagenet" , input_shape=(300 , 300 ,3)  )
```

```python
t_model.summary()
```

```python
for layer in t_model.layers:
  layer.trainable = False
```

```python
t_model.summary()
```

```python
from keras.layers import Dense , Flatten
```

```python
x = Flatten()(t_model.layers[-1].output)
h = Dense (units = 128 , activation= "relu" , kernel_initializer="he_uniform")(x)
o = Dense(units=1 , activation="sigmoid" )(h)
```

```python
from keras.models import Model
```

```python
m = Model(inputs =t_model.input , outputs =o )
```

```python
m.compile(loss = "binary_crossentropy" , optimizer="adam" , metrics=["accuracy"])
```

```python
m.fit(train_ganator , epochs=10)
```

```python
m.evaluate(test_ganator)
```
